```python
class Employee:
    salary = "12"
    # Dunder Methode which is automatically called
    
    def __init__(self, name:str, salary:int):
        print("Created")
        self.name = name
        self.salary = salary
        
    
    @staticmethod
    def greet():
        print(f"Good Morning")
```

```python
name= Employee("Shivpratap",1200)
print(name.name)
#Employee.greet(name)
```

```python
class Calculator:
    def __init__(self):
        pass
    def addNum(self , a,b):
        return a+b
    def power(self,a,b):
        return a**b
    def squareRoot(self,a):
        return a**(1/2)
    
```

```python
cal = Calculator()
print(cal.squareRoot(25))
```
