```python
'''Write a recursive function to find factorial of n.

Input: n = 5
Output: 120

Explanation: 5! = 5 × 4 × 3 × 2 × 1 = 120'''

def fact(n):
    if n == 1 or n == 0:
        return 1
    return n*fact(n-1)

print(fact(5))
```

```python
'''Write a recursive function to find sum of first n natural numbers.

Input: n = 5
Output: 15

Explanation: 1 + 2 + 3 + 4 + 5 = 15'''

# sun(n) = n + (n-1) + (n-2) + ....

def sum(n):
    if n == 1:
        return 1
    return n + sum(n-1)
print(sum(10))
```

```python
'''Write a recursive function to find the nth Fibonacci number.

Input: n = 6
Output: 8

Explanation: Sequence: 0, 1, 1, 2, 3, 5, 8...'''

''' 
fab(6): 0 1 2 3 4 5 6 
'''
def fibonacci(n):
    # Base Cases
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # Recursive Case
    return fibonacci(n - 1) + fibonacci(n - 2)


# Testing
print("Fibonacci Sequence:")
for i in range(7):
    print(f"fib({i}) = {fibonacci(i)}")

'''
fibonacci(6)
    │
    ├── fibonacci(5)
    │       │
    │       ├── fibonacci(4)
    │       │       │
    │       │       ├── fibonacci(3)
    │       │       │       │
    │       │       │       ├── fibonacci(2)
    │       │       │       │       │
    │       │       │       │       ├── fibonacci(1) → returns 1
    │       │       │       │       └── fibonacci(0) → returns 0
    │       │       │       │       └── returns 1 + 0 = 1
    │       │       │       │
    │       │       │       └── fibonacci(1) → returns 1
    │       │       │       └── returns 1 + 1 = 2
    │       │       │
    │       │       └── fibonacci(2) → returns 1
    │       │       └── returns 2 + 1 = 3
    │       │
    │       └── fibonacci(3) → returns 2
    │       └── returns 3 + 2 = 5
    │
    └── fibonacci(4) → returns 3
    └── returns 5 + 3 = 8 ✓
'''
```

```python
'''
Write a recursive function to count the number of digits in a number.

Input: n = 12345
Output: 5
'''
def count(n):
    n = abs(n)
    if n < 10:
        return 1
    else:
        return 1 + count(n//10)

print(count(123456789))
    
'''
count_digits(12345)
│
├── Is 12345 < 10? NO
├── Return 1 + count_digits(12345 // 10)
│            = 1 + count_digits(1234)
│                  │
│                  ├── Is 1234 < 10? NO
│                  ├── Return 1 + count_digits(1234 // 10)
│                  │            = 1 + count_digits(123)
│                  │                  │
│                  │                  ├── Is 123 < 10? NO
│                  │                  ├── Return 1 + count_digits(123 // 10)
│                  │                  │            = 1 + count_digits(12)
│                  │                  │                  │
│                  │                  │                  ├── Is 12 < 10? NO
│                  │                  │                  ├── Return 1 + count_digits(12 // 10)
│                  │                  │                  │            = 1 + count_digits(1)
│                  │                  │                  │                  │
│                  │                  │                  │                  ├── Is 1 < 10? YES ✓
│                  │                  │                  │                  └── Return 1
│                  │                  │                  │
│                  │                  │                  └── Return 1 + 1 = 2
│                  │                  │
│                  │                  └── Return 1 + 2 = 3
│                  │
│                  └── Return 1 + 3 = 4
│
└── Return 1 + 4 = 5 ✓
'''
```

```python
'''
Write a recursive function to find sum of digits of a number.

Input: n = 1234
Output: 10

Explanation: 1 + 2 + 3 + 4 = 10
'''

def sumDigits(n):
    n = abs(n)
    def helper(n):
        if n < 10:
            return n
        d = n%10
        return d + helper(n//10)
    return helper(n)


print(sumDigits(111230))

'''
sumDigits(1234)
│
├── d = 4, return 4 + sumDigits(123)
│                     │
│                     ├── d = 3, return 3 + sumDigits(12)
│                     │                     │
│                     │                     ├── d = 2, return 2 + sumDigits(1)
│                     │                     │                     │
│                     │                     │                     └── return 1
│                     │                     │
│                     │                     └── return 2 + 1 = 3
│                     │
│                     └── return 3 + 3 = 6
│
└── return 4 + 6 = 10 ✓

1 + 2 + 3 + 4 = 10 ✓
'''
```

```python
'''
Write a recursive function to reverse a string.

Input: "hello"
Output: "olleh"
'''

def reverse(string):
    if len(string) <= 1:
        return string
    return string[-1] + reverse(string[:-1])

print(reverse("shivpratap"))  # Output: patarpvihs
print(reverse("hello"))       # Output: olleh
print(reverse("a"))           # Output: a
print(reverse(""))            # Output: (empty)
print(reverse("ab"))          # Output: ba
print(reverse("hello world")) # Output: dlrow olleh
print(reverse("12345"))
```
