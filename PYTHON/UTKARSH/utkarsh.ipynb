```python
#import datetime

#b= datetime.datetime.now()
#print(b)
from datetime import datetime

# Define two datetime objects
date1 = datetime(2026,6,15, 11, 30, 0)  # Jan 15, 2025, 10:00 AM
date2 = datetime(2026, 2, 11, 12, 0, 0)   # Jan 15, 2024, 9:00 AM

# Subtract the dates
# The result is a timedelta object
time_difference = date1 - date2

# Print the result
print(f"Time difference: {time_difference}")

# Access components of the timedelta
print(f"Days: {time_difference.days}")
print(f"Seconds (total): {time_difference.total_seconds()}")


```
