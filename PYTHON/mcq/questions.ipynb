```python
def swapAdjecent(l:list)->list:
    for i in range(0,len(l)-1,2):
        l[i],l[i+1]=l[i+1],l[i]
    return l
print(swapAdjecent([1,2,3,4,5,6,7,7]))
```

```python
def removeElement(l:list,el:int)->list:
    for i in l:
        if i==el:
            l.remove(i)
            
    return l

print(removeElement([1,2,3,4,5,6,7,8],7))
```

```python

def joinStringWithSpecificChar(word:str, iterable: list |tuple | str | dict)->str:
    newStr = ""
    for i in iterable:
        if i != iterable[-1]:
            newStr += i+word
        else:
            newStr += i
    return newStr

oders = {
    "zebra":7,
    "apple":5,
    "milk":2
}
result = []
for i in sorted(oders, reverse=True):
    result.append(i[-1]+str(oders[i]))
print(joinStringWithSpecificChar(",",result))
```

```python
def shiv(word:str, iterable: list |tuple | str | dict)->str:
    newStr = ""
    for i in iterable:
        if i != iterable[-1]:
            newStr = newStr+i+word
        else:
            newStr += i
    return newStr

print(shiv(",", ["2","4","5"]))
```
