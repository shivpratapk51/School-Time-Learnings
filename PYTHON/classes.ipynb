# Classes #

```python
class Human():
    def __init__(self, n, o):
        # print(n, o)
        self.name = n
        self.occupation = o
    
        
    def __repr__(self):
        return (f"Human(Name = {self.name}, Occupation = {self.occupation})")
    
    def do_work(self):
        if self.occupation == "Student":
            print(f"{self.name} is a student")
        elif self.occupation == "Teacher":
            print(f"{self.name} is a {self.occupation}")
        else:
            print("Not Given Occupation")
            
    def speaks(self):
        if self.occupation == "Teacher":
            print(f"{self.name}: How are you students? ")
        elif self.occupation == "Student":
            print(f"{self.name}: How are you teacher?")
        else:
            print("How are you guys?")
```

```python
shiv = Human("Shivpratap", "Student")
teacher = Human("Jitendra", "Teacher")
teacher.do_work()
teacher.speaks()
```

# Inheritance #

* Inheritance means a master class which have some sort of own properties and method but sometimes we need a same properties of master class but in different ways or sometime we need to add more properties or delete properties, then we define a sub class which is inherit from master class but having similar properties but in different way.

```python
class Vehicle:
    def general_usage(self):
        print("General Use: Transportation")
    def __init__(self):
        pass
    
class Car(Vehicle):
    def __init__(self):
        print("I'm a car.")
        self.wheels = 4
        self.has_roof = True
    def specific_usage(self):
        print("Specific Usage: commute to work, vacation with family")
    

class MotorCycle(Vehicle):
    def __init__(self):
        print("I'm a motorcycle.")
        self.wheels = 2
    
    def specific_usage(self):
        print("Specific Usage: racing, road trip")
```
