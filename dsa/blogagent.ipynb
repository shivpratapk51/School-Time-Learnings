```python
from __future__ import annotations

import operator
from typing import TypedDict, List, Annotated

from pydantic import BaseModel, Field
from langgraph.graph import StateGraph, START, END
from langgraph.types import Send

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
class Task(BaseModel):
    id: int
    title: str
    brief: str = Field(..., description="What to cover?")
```

```python
class Plan(BaseModel):
    blog_title: str
    tasks: List[Task]
```

```python
class State(TypedDict):
    topic:str
    plan:Plan
    #reducer: results from worker get concatented autometically
    sections: Annotated[List[str], operator.add]
    final: str
```

```python
def lenght(a: int, b: int) -> None:
    count = 0
    if a<b:
        for i in range(a, b+1):
            # print(i)
            count += 1
        print(f"Number of digits from {a} to {b} is {count}.")
    else:
        print("first num should be less than second num")

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
lenght(a,b)
```

```python
import instaloader

def downloadReel(url: str):
    l = instaloader.Instaloader()
    
    try:
        shortCode = url.split("/")[-2]
        post = instaloader.Post.from_shortcode(l.context, shortCode)
        print(f"Downloading reel from: {post.owner_username}...")
        l.download_post(post, target=f"{post.owner_username}_reel")
        print("Download Completed!!")
    except Exception as e:
        print(f"Error: {e}")
        
url = input("Enter url: ")
downloadReel(url)
```
